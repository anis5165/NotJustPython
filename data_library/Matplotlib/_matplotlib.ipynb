```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
x = ['Mon','Tues','Wed','Thurs','Fri']
y = [10,20,15,25,12]

plt.plot(x,y)
plt.title("Bakery Sales This Week")
plt.xlabel("Days of the Week")
plt.ylabel("Number of Sales per day")

plt.show()
```

```python
months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months,sales, color='green', linestyle='-.', marker='o', label='2025 sales data')
plt.title("2025 Sales Data Report")
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.legend(loc='lower right', fontsize=10)
plt.grid(color='gray', linestyle=':', linewidth=1)
#x and y range control using xlim and ylim.
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
#bar chart
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.bar(product, sales, color='orange', label='Sales 2025') #for horizontal use barh
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.title("Product sales comparison")
plt.show()
```

```python
#pie graph
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'])
plt.title("Revenue Contribution By Region")
plt.show()
```

```python
#histogram

Scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(Scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Distribution of Students')
plt.show()
```

```python
#scattered chart

hours_study = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_study,exam_scores, color='red', marker='^', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title("Relationship between study time and exam score")
plt.legend()
plt.grid()
plt.show()


```

```python
#multiple scattered chart

plt.scatter([1,2,3],[50,60,70], color='blue', label='Class A')
plt.scatter([1,2,3],[45,50,69], color='red', label='Class B')


plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title("Relationship between study time and exam score")
plt.legend()
plt.grid()
plt.show()
```

```python
#subplots

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) #1 row, 2 column, 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) #1row, 2column, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()

```

```python
#subplot

fig, ax = plt.subplots(1,2, figsize=(10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y, color='pink')
ax[0].set_title("Line chart")

ax[1].bar(x,y,color='red')
ax[1].set_title("Bar chart")

fig.suptitle('Comparison of lie and bar charts')

plt.tight_layout() #use only just before plt.show() 
plt.show()
```

```python
#save file

x = [1,2,3,4]
y = [10,20,15,25] 

plt.plot(x,y, color='blue', marker='o')
plt.title("Simple Line Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.savefig('line_chart.pdf', dpi=300, bbox_inches='tight') #savefig should be used before plt.show() #save in png, jpg, svg, pdf etc.
plt.show()
```
