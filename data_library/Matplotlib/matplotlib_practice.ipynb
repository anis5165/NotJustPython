```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("students.csv")
print(df)
```

```python
# Bar Chart: Compare Math scores of each student.
plt.figure(figsize=(15,5))
plt.bar(df["Student"],df["Math"], label="Student Math Score")
plt.title("Student Math Score Chart")
plt.xlabel("Student")
plt.ylabel("Math Score")
plt.tight_layout()
plt.savefig("Student Math Score Chart.png", dpi=300)
plt.show()
```

```python
# Line Chart: Plot the trend of Math, Science, and English scores for all students.
plt.figure(figsize=(10,6))

# plt.plot(df["Student"], df[["Math","Science","English"]], marker="o", label="The trend of Math, Science and English Scores")

                                # OR 
                                
#full custumization when use different-2 lines
plt.plot(df["Student"], df["Math"], marker="o", color="blue", label="Math Scores")
plt.plot(df["Student"], df["Science"], marker="o", color="green", label="Science Scores")
plt.plot(df["Student"], df["English"], marker="o", color="red", label="English Scores")

plt.title("The trend of Math, Science, English Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
# Stacked Bar Chart: Show total academic performance (Math + Science + English) for each student.
total_score = df[["Math","Science","English"]].sum(axis=1) # axis 1 defines rows wise

plt.figure(figsize=(10,5))
plt.barh(df["Student"],total_score, color="purple", label="Total academic performance")
plt.title("Total academic performance (Math + Science + English) for each student")
plt.xlabel("Scores")
plt.ylabel("Students")
plt.legend()
plt.show()
```

```python
# Scatter Plot: Plot Hours_Studied vs. Math scores. What kind of correlation do you see?

plt.figure(figsize=(10,5))
plt.scatter(df["Hours_Studied"],df["Math"], color="green", marker="^", label="Hours_Studied vs. Math scores")
plt.title("Hours_Studied vs. Math scores Scatter chart")
plt.xlabel("Hours Studied")
plt.ylabel("Math Scores")
plt.legend()
plt.grid()
plt.show()
```

```python
# Histogram: Show the distribution of Science scores./

plt.figure(figsize=(10,6))
plt.hist(df["Science"], bins=5, edgecolor="black")
plt.title("The distribution of Science Score")
plt.xlabel("Science Score")
plt.ylabel("Students")
plt.show()
```

```python
# Pie Chart: Show percentage contribution of Sports_Score for each student.

plt.figure(figsize=(10,6))
plt.pie(df["Sports_Score"], labels=df["Student"], autopct="%1.1f%%", startangle=90)
plt.show()
```

```python
# Show percentage contribution of maths for each student.
plt.figure(figsize=(10,6))
plt.pie(df["Math"], labels=df["Student"], autopct="%1.1f%%", startangle=90, shadow=True)
plt.show()
```

```python
# Boxplot: Show variation in English scores.

plt.figure(figsize=(10,6))
plt.boxplot(df["English"], patch_artist=True)
plt.ylabel("Score")
plt.title("The variation in English scores")
plt.show()
```

```python
# Subplots: Create a figure with 4 subplots (Line, Bar, Histogram, Scatter).

fig, ax = plt.subplots(1,4, figsize=(16,5))


ax[0].plot(df["Student"], df["English"], linestyle="-." ,marker='o', label="English score")
ax[0].set_title("English Score")
ax[0].set_xlabel("Student")
ax[0].set_ylabel("English Score")
ax[0].legend()
ax[0].grid()




ax[1].bar(df["Student"], df["Math"], color="green", label="Math Score")
ax[1].set_title("Math Score")
ax[1].set_xlabel("Student")
ax[1].set_ylabel("Math Score")
ax[1].legend()


ax[2].hist(df["Science"], bins=5, edgecolor="black", color="red")
ax[2].set_title("Distribution of Science Score")
ax[2].set_xlabel("Science score")
ax[2].set_ylabel("Student")


ax[3].scatter(df["Student"],df["Hours_Studied"], color="blue", marker="^", label="Hours studied")
ax[3].set_title("Student v/s Hour_studied")
ax[3].set_ylabel("Hour studied")
ax[3].set_xlabel("Student")
ax[3].grid()



plt.show()
```

```python
# Customized Plot: Create a scatter plot with student names as annotations.

plt.figure(figsize=(10,5))
plt.scatter(df["Hours_Studied"], df["Math"], color="blue", marker="o")

for i, name in enumerate(df["Student"]):
    plt.text(df["Hours_Studied"][i] + 0.2, df["Math"][i] + 0.2, name)

plt.title("Hours Study vs Maths Score")
plt.xlabel("Hours Studied")
plt.ylabel("Maths Scores")
plt.grid()
plt.show()
```

```python
# Advanced: Compare average academic scores vs. Sports_Score in a grouped bar chart.
import numpy as np

df["Avg_Academic_score"] = df[["Math","Science","English"]].mean(axis=1)

bar_width = 0.35
index = np.arange(len(df["Student"]))

plt.bar(index,df["Avg_Academic_score"], bar_width, label="Avg Academic Score", color="blue")
plt.bar(index + bar_width, df["Sports_Score"], bar_width, label="Sports Score", color="orange")

plt.title("Avg Academic Score vs Sports Score")
plt.xlabel("Student")
plt.ylabel("Scores")
plt.xticks(index + bar_width / 2, df["Student"])
plt.legend()
plt.show()
```
