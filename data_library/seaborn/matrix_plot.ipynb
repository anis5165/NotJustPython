## Matrix plot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the example tips dataset
gap = px.data.gapminder()
print(gap)
```

```python
#heat map - axes level function, there is not figure function is used to plot head map 
temp_df = gap.pivot(index="country", columns="year",values="lifeExp")

plt.figure(figsize=(15,10)),
sns.heatmap(data=temp_df)
```

```python
#annot and linewidths and cmap to choose color
temp_df2 = gap[gap["continent"] == "Europe"].pivot(index="country", columns="year",values="lifeExp")

# print(temp_df2)
plt.figure(figsize=(15,10))
sns.heatmap(data=temp_df2, annot=True,linewidths=0.5, cmap="summer") #annot=true means it disply data in text and linewidths provide gap
```

```python
#clustermap
#plot a matrixx dataset as a hierarchically-clustered heatmap.
#This function requires scripy to be available
iris = px.data.iris()
iris
```

```python
cluster = sns.clustermap(iris.iloc[:,[0,1,2,3]])
print(cluster)
```
