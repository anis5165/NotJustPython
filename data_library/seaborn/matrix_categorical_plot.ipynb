## Matrix plot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the example tips dataset
gap = px.data.gapminder()
print(gap)
```

```python
#heat map - axes level function, there is not figure function which plot head map 
temp_df = gap.pivot(index="country", columns="year",values="lifeExp")

plt.figure(figsize=(15,10)),
sns.heatmap(data=temp_df)
```

```python
#annot and linewidths and cmap to choose color
temp_df2 = gap[gap["continent"] == "Europe"].pivot(index="country", columns="year",values="lifeExp")

# print(temp_df2)
plt.figure(figsize=(15,10))
sns.heatmap(data=temp_df2, annot=True, linewidths=0.5, cmap="viridis") #annot=true means it disply data in text and linewidths provide gap
```

```python
#clustermap
#plot a matrixx dataset as a hierarchically-clustered heatmap.
#This function requires scripy to be available
iris = px.data.iris()
iris
```

```python
cluster = sns.clustermap(iris.iloc[:,[0,1,2,3]])
print(cluster)
```

## Categorical plot

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
#categorical plot - categorical scatter plot
#1. Stripplot
#2. swarmplot
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
#strip plot
sns.stripplot(data=tips, x="day", y="total_bill") #jitter=false
```

```python
#catplot - figure level function
sns.catplot(data=tips, x="day", y="total_bill", kind="strip")
```

```python
#jitter
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", jitter=0.3, hue="sex") #jitter = false means every dot in a single line
```

```python
# swarplot
sns.swarmplot(data=tips, x="day", y="total_bill")
```

```python
# swarplot using figure level function 
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
#true parameter
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm", hue="sex")
```

```python
#categorial distribution plots
#1.boxplot - A box plot is a standardized way of displaying the distribution of data based on A 5 number summary
# ("minimum, maximum, first quartile[Q1], third quartile[Q3], median"). it can tell you about your outliers and what their values are.
#Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if grouped and if and how your data is skewed.
#2.violinplot
```

```python
#boxplot
# sns.boxplot(data=tips, x="day",y="total_bill")
sns.boxplot(data=tips, x="sex",y="total_bill")
```

```python
# using figure level function 
sns.catplot(data=tips, x="day", y="total_bill",kind="box")
```

```python
#hue paramter
sns.boxplot(data=tips, x="day",y="total_bill", hue="sex")
```

```python
#single boxplot -> numerical col
sns.boxplot(data=tips, y="total_bill")
```

```python
#violinplot = (boxplot + KDEplot)
sns.violinplot(data=tips, x="day", y="total_bill")
```

```python
#using figure level function voilinplot
sns.catplot(data=tips, x="day", y="total_bill", kind="violin", hue="sex", split=True)
```

```python
#categorical estimate plot -> for central tendency
#1. Barplot
#2. Pointplot
#3. Countplot
```

```python
#barplot
sns.barplot(data=tips, x="sex", y="total_bill")
```

```python
#hue parameter
sns.catplot(data=tips, x="sex", y="total_bill", kind="bar", hue="smoker")
```

```python
#estimate
import numpy as np
sns.barplot(data=tips, x="sex",y="total_bill", hue="smoker", estimator=np.max)
```

```python
#average total bill of both gender
sns.barplot(data=tips, x="sex", y="total_bill", color="red", errorbar=None)
```

```python
#pointplot

sns.pointplot(data=tips, x="sex",y="total_bill", hue="smoker")
```

```python
#with errorbar=None
sns.pointplot(data=tips, x="sex",y="total_bill", hue="smoker", errorbar=None)
```

```python
#countplot - A special case for bar plot is when you want to show the number of observations in each category rather than 
# computing a statistic for a second variable this is similar to histogram over A categorical rather than quantitative variable

sns.countplot(data=tips, x="sex")
```

```python
# with hue parameter 
sns.countplot(data=tips, x="sex", hue="day")
```

```python
#faceting using caplot
sns.catplot(data=tips, x="sex", y="total_bill", col="smoker", kind="box", row="time")
```
