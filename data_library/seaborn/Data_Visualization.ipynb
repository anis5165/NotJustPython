### We use distribution plots to visualize the distribution of quantitative data
### Let's discuss some plots that allow us to visualize the distribution of a data set these plots are:
### * distplot/histplot
### * jointplot
### * pairplot
### * rugplot


```python
df = sns.load_dataset("tips")
```

```python
df["size"].unique()
```

```python
# displot/histplot 
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"], kde=True)
```

```python
# jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter") # kind = "reg", "hex", "kde", "resid"
```

```python
# pairplot
sns.pairplot(df, hue="sex", palette="coolwarm")
```

```python
# rugplot
sns.rugplot(df["total_bill"])
```

```python
#categoria; plots
# countplot
# boxplot
# violinplot
# stripplot
# swarmplot
# barplot
```

```python
#countplot
sns.countplot(x="sex", hue="smoker", data=df)
```

```python
#boxplot
sns.barplot(x="sex", y="tip", data=df, estimator=np.sum) #estimator can be np.mean, np.median, np.std
```

```python
#boxplot
sns.boxplot(x="tip",y="day", data=df, palette="rainbow")
```

```python
# Box plot and Viral plot -  Box plots and viral plots are used to shawn the distribution of categorical data a box plot shows 
# the distribution of quantitative data in a way that facilities comparison between variables or across levels of categorical
# variable the box shows the coratiles of the data while the whiskers extend to show the rest of the distribution except for 
# points that are determined to be outliers using a method that is function of the interquartile range
sns.violinplot(x="day", y="total_bill", data=df)
```

```python
#stripplot
sns.stripplot(x="day",y="total_bill", data=df)
```

```python
#swanplot
sns.swarmplot(x="day",y="total_bill", data=df)
```

```python
#violinplot + swarmplot
sns.violinplot(x="tip", y="day",data=df)
sns.swarmplot(x="tip", y="day",data=df)
```

```python
#matrix plot - headmap

flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips_corr = tips[["total_bill", "tip", "size"]]
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr(), annot=True)
```

```python
#clustermap
sns.clustermap(tips_corr.corr())
```

```python
pivot_flight = flights.pivot_table(index="month", columns="year", values="passengers")
pivot_flight
```

```python
# pivot table heatmap 
sns.heatmap(pivot_flight)
```

```python
#regression plot - lmplot
sns.lmplot(x = "total_bill", y="tip", data=tips, hue="sex", palette="coolwarm", markers=["^","v"], scatter_kws={'s':100})
```
