## Regression plot

```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
#regplot nad lmplot - In the simplest invocation both functions draw a scatter plot of two variables X and Y and then fit the regression
#  model Y ~ X and plot the resulting regression line and a 95% confidence interval for that regression


#in regplot hue parameter is not available
sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
#lmplot
#in lmplot we use use parameter
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
#residplot
sns.residplot(data=tips, x="total_bill", y="tip")
```

## Multigrid plot

```python
#A second way to plot facet plots -> facetgrid
# figure level -> relplot, displot, catplot, lmplot

sns.catplot(data=tips, x="sex", y="total_bill", kind="violin", col="day", row="time")
```

```python
g = sns.FacetGrid(data=tips, col="day", row="time", hue="smoker")
g.map(sns.violinplot,"sex","total_bill")
g.add_legend()
```

```python
#pairplot
iris = px.data.iris()

iris
```

```python
#pairgrid
sns.pairplot(iris, hue="species")
```

```python
#pairgrid
g = sns.PairGrid(data=iris, hue="species")
#g.map
g.map(sns.scatterplot)
```

```python
#map_diagonal -> map_offdiagonal
g = sns.PairGrid(data=iris, hue="species")
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
#map_diagonal -> map_upper -> map_lower

g = sns.PairGrid(data=iris, hue="species", vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
#jointplot
sns.jointplot(data=tips,x="total_bill",y="tip", kind="hist", hue="sex") #kind = "scatter", "kde", "hist", "hex", "reg", "resid"
```

```python
#jointgrid
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.kdeplot,sns.violinplot) 
```
