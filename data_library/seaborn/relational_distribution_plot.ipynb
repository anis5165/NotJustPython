```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
#scatter plot - axis level function
scatter = sns.scatterplot(data=tips, x='total_bill', y="tip", hue="sex", style="time", size="size")
print(scatter)
```

```python
#relpot - figure level function - shape square
#in figure level function we cannot set plot size using plt.figure() because both are points different figures.
# but if you want to increase heigth [use height and aspect property in figure level functions]
sns.relplot(data=tips, x='total_bill', y="tip", kind='scatter',hue="sex", style="time",size="size", height=7, aspect=2)
```

```python
#line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India'] 
print(temp_df)
```

```python
lineplot = sns.lineplot(data=temp_df, x="year", y="lifeExp", marker="o")
print(lineplot)
```

```python
relpot = sns.relplot(data=temp_df, x="year", y="lifeExp", kind="line", marker="o")
print(relpot)
```

```python
temp_df = gap[gap["country"].isin(["India","China","Pakistan"])]
print(temp_df)
```

```python
#in relplot - hue(legend) always made outside the plot, because relplot is a figure level function it has a access of outside the plot
multiline = sns.relplot(data=temp_df, kind="line", x="year", y="lifeExp", hue="country") #CPU dispatcher tracer already initlized
print(multiline)
```

```python
temp_df2 = gap[gap["country"].isin(["India","Germany","Brazil"])]
# print(temp_df)
linegraph = sns.relplot(data=temp_df2, x="year", y="lifeExp", kind="line", hue="country", style="continent", size="continent")
print(linegraph)
```

```python
#facet plot - to make multiple plot on a particular column e.g column gender means 1 plot for male and another for female
#facet plot -> only work with figure level function means relplot and i will not work with scatterplot and lineplot


# facetplot = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="sex")
# facetplot = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="smoker" ,row="sex")
# facetplot = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="day" ,row="sex")
facetplot = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="sex")
```

```python
#column wrap
scatterplot = sns.relplot(data=gap, x="lifeExp", y="gdpPercap", kind="scatter", col="year", col_wrap=3)
print(scatterplot)
```

# Distribution plot

```python
#figure level function  -> displot
#axis level -> histplot, kdeplot, rugplot
print(tips)
```

```python
#histplot
histplot = sns.histplot(data=tips, x="total_bill")
print(histplot)
```

```python
#displot - figure level function
distplot = sns.displot(data=tips, x="total_bill", kind="hist")
print(distplot)
```

```python
#bins parameter
distplot = sns.displot(data=tips, x="total_bill", kind="hist", bins=5)
print(distplot)
```

```python
#It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

distplot = sns.displot(data=tips, x="day", kind="hist")
print(distplot)
```

```python
#hue(legend) parameter 
distplot = sns.displot(data=tips, x="tip", hue="sex", kind="hist")
print(distplot)
```

```python
#element -> step to clear view(optional)
distplot = sns.displot(data=tips, x="tip", hue="sex", element="step",kind="hist")
print(distplot)
```

```python
titanic = sns.load_dataset("titanic")
print(titanic)
```

```python
distplot = sns.displot(data=titanic, x="age", hue="sex", element="step",kind="hist") #count b/w male and female using draph
print(distplot)
```

```python
#faceting using col and row
distplot = sns.displot(data=tips, x="tip", col="sex", element="step",kind="hist")
print(distplot)
```

```python
#kdeplot - Kernel Density Estimate
#Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernal, producing a continuous density estimate

kdeplt = sns.kdeplot(data=tips, x="total_bill")
```

```python
#using figure level function
sns.displot(kind="kde", data=tips, x="total_bill")
```

```python
#hue -> fill
sns.displot(kind="kde", data=tips, x="total_bill", hue="sex", fill=True)
```

```python
#Rugplot
# Plot marginal distribution by drawing ticks along the x and y axes.
#This function is intended to complement other plots bu shawing the location of individual oberservation in an unobtrusive way.
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
```

```python
#Bivariate histogram
# a bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x="total_bill", y="tip")
sns.displot(kind="hist" ,data=tips, x="total_bill", y="tip")
```

```python
#Bivariate Kdeplot
# a bivariate KDE plot smooths the (x,y) observations with a 2D Gaussian

sns.kdeplot(data=tips, x="total_bill", y="tip")
```
